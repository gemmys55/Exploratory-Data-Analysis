# Providing a dataset of employees working in ABC company. It consists of 458 rows and 9 columns. The company needs the detailed report and explanation of their employees in each team, also need to identify the following

## 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
df['Team'].value_counts()
```

```python
team_counts = df['Team'].value_counts()

# Percentage split with respect to the total employees
percentage_split = team_counts / len(df) * 100

# Display the counts and percentages
team_info = pd.DataFrame({'Team Count': team_counts, 'Percentage Split': percentage_split})
print(team_info)
```

```python
import matplotlib.pyplot as plt

colors = ['steelblue', 'red', 'green', 'yellow', 'skyblue']
plt.bar(team_counts.index,team_counts.values,color=colors)
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.title('Percentage Splitting of Employees w.r.t each Team')
plt.xticks(rotation='vertical')
plt.show()
```

#### By analysing the above bar graph , it is clear that the team "New Orleans Pelicans" has more percentage split compared with other teams

### 3.Find from which age group most of the employees belong to.

```python
df['Age']
```

```python
age_counts = df['Age'].value_counts()
most_common_age = age_counts.idxmax()

print(f"The age group with the most employees is around {most_common_age} years.")
```

```python

df['Age Group'] = df['Age'].apply(lambda age:'20-25' if 20 <= age <= 25 else ('26-30' if 26 <= age <= 30 else ('31-35' if 31 <= age <= 35 else '36 and above')))
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Create a bar chartadf['Age Group'].value_counts()
age_group_counts=df['Age Group'].value_counts()
plt.hist(df['Age'],color='yellow',edgecolor='red')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Individuals')
plt.show()
```

### 4.Find out under which team and position, spending in terms of salary is high

```python
import pandas as pd

# Assuming 'df' is your DataFrame with columns 'Team', 'Position', and 'Salary'
# Replace 'Salary' with the actual column name for salary in your DataFrame
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum()

max_spending_team_position = team_position_salary.idxmax()
max_spending_value = team_position_salary.max()

print(f"The team and position withs spending in terms of salary is high {max_spending_team_position} with a total spending of {max_spending_value}.")
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with columns 'Salary' and 'Age'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='blue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salary vs Age')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with a 'Salary' column
plt.figure(figsize=(12, 8))
sns.boxplot(x='Team', y='Salary', data=df, palette='Set3')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.title('Boxplot of Salary by Team')
plt.show()
```

### 5.Find if there is any correlation between age and salary , represent it visually.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'Salary']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Age and Salary')
plt.show()

```
